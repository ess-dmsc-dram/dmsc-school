```python
from matplotlib import pyplot as plt

# Set the style to xkcd
plt.xkcd()

bar_width = 0.2
x_2023 = [1 - bar_width, 2 - bar_width, 3 - bar_width, 4 - bar_width, 5 - bar_width]
x_2024 = [1 + bar_width, 2 + bar_width, 3 + bar_width, 4 + bar_width, 5 + bar_width]

# Hardcoded data for the years here
overall_satisfactions = {
    2023: [0, 0, 0, 2, 4],
    2024: [0, 0, 0, 7, 7],
    2025: [0, 0, 0, 0, 0],  # TODO: Add data for 2025
}
number_of_participants = {
    2023: 13,
    2024: 20,
    2025: 0,  # TODO: Add data for 2025
}

# Make x-axis labels
x_satisfaction = [1, 2, 3, 4, 5]
x_axis_offset = bar_width * (len(overall_satisfactions) / 2 - 0.5)
x_overall_satisfaction = {
    year: [(x - x_axis_offset) + (i_year * bar_width) for x in x_satisfaction]
    for i_year, year in enumerate(overall_satisfactions)
}

for year, satisfaction in overall_satisfactions.items():
    # Plot the data
    num_participants = number_of_participants[year]
    plt.bar(
        x_overall_satisfaction[year],
        satisfaction,
        alpha=0.7,
        label=f"{year} ({sum(satisfaction)} responses / {num_participants})",
        width=bar_width,
    )

plt.ylim(-0.8, 10)
plt.title("Overall Satisfaction of the DMSC Summer School")
plt.xlabel("Rating")
plt.ylabel("Number of Responses")
plt.legend()
```
